### Data Cleaning

```python
#Checking for missing values
df.isnull().sum()
```

```python
# dropping nan and unnecessary columns
df.drop(['ind1', 'ind2', 'Row ID+O6G3A1:R6', 'Order ID', 'Customer ID', 'Customer Name','Product ID','Ship Date','Region'], axis=1, inplace=True)
```

```python
# checking for duplicates // no duplicates
df_dups= df[df.duplicated()]
df_dups
```

```python
# Creatinh year and month columns
df['Order Date']=pd.to_datetime(df['Order Date'],format='%d-%m-%Y')
df['Year'] = df['Order Date'].dt.year
df['Month_Number'] = df['Order Date'].dt.month
df['Month_Name'] = df['Order Date'].dt.month_name()
```
